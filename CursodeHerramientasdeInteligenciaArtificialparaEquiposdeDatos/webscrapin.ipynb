```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la página de noticias de TechCrunch
url = 'https://techcrunch.com/'

# Realizar la solicitud HTTP
response = requests.get(url)

# Comprobar que la solicitud fue exitosa
if response.status_code == 200:
    # Analizar el contenido HTML
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Listas para almacenar los datos
    titles = []
    links = []
    authors = []
    dates = []
    images = []

    # Encontrar todas las noticias en la página
    articles = soup.find_all('article')
    
    for article in articles:
        # Extraer el título
        title_tag = article.find('h1', class_='wp-block-post-title')
        title = title_tag.get_text(strip=True) if title_tag else 'No title'
        
        # Extraer el enlace
        link_tag = article.find('a', href=True)
        link = link_tag['href'] if link_tag else 'No link'
        
        # Extraer el autor
        author_tag = article.find('div', class_='wp-block-tc23-author-card-name')
        author = author_tag.get_text(strip=True) if author_tag else 'No author'
        
        # Extraer la fecha
        date_tag = article.find('time')
        date = date_tag.get_text(strip=True) if date_tag else 'No date'
        
    
        
        # Añadir los datos a las listas
        titles.append(title)
        links.append(link)
        authors.append(author)
        dates.append(date)

    
    # Crear un DataFrame de pandas
    df = pd.DataFrame({
        'Title': titles,
        'Link': links,
        'Author': authors,
        'Date': dates,
    
    })
    
    # Mostrar el DataFrame
    print(df)
    
    # Guardar el DataFrame a un archivo CSV
    df.to_csv('techcrunch_news.csv', index=False)
else:
    print(f'Error al acceder a la página. Código de estado: {response.status_code}')
```
