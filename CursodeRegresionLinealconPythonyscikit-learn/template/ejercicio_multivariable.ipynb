# Regresión lineal multivariable

## Importando datos

```python
import pandas as pd
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data',
            header=None, sep=r'\s+')
df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df.head()
```

## Creando modelo con sklearn

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

X = df[['RM', 'INDUS']].values
y = df['MEDV'].values.reshape(-1, 1)
sc_x = StandardScaler()
sc_y = StandardScaler()
X_std = sc_x.fit_transform(X)
y_std = sc_y.fit_transform(y)

slr = LinearRegression()
slr.fit(X_std, y_std)
```

## Análisis de resultados

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import *# o Axes3D
import numpy as np 
import pandas as pd

x1_range = np.arange(df['RM'].min(), df['RM'].max(), 0.1)
x2_range = np.arange(df['INDUS'].min(), df['INDUS'].max(), 0.1)

X1, X2 = np.meshgrid(x1_range, x2_range)

plano = pd.DataFrame({'RM': X1.ravel(), 'INDUS': X2.ravel()})
pred = slr.predict(plano)

# Asegurarse que pred esté en la forma correcta para inverse_transform
pred = sc_y.inverse_transform(pred.reshape(-1, 1)).reshape(X1.shape)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, pred, alpha=0.4)

ax.scatter3D(df['RM'], df['INDUS'], df['MEDV'], color='r', marker='.')
ax.view_init(elev=10, azim=5)
plt.show()
```

## Generando predicción con el modelo

```python
num_rooms_std = sc_x.transform(np.array([5.0]).reshape(-1, 1))
price_std = slr.predict(num_rooms_std)
print('El precio en miles es de %.3f'%
      sc_y.inverse_transform(price_std))
```
