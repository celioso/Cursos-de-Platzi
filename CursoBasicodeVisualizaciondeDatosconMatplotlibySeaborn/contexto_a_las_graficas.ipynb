```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2,figsize=(5,5))
axes[0].plot(x,y,label="$sin(x)$")
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
fig.tight_layout() 
```

```python
plt.plot(x,y,label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
fig.tight_layout() 
```

### C√≥mo dar contexto a los gr√°ficos
#### Creamos nuestras gr√°ficas comunes

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b')
ax2.plot(y,x, 'r')
plt.show()
fig.tight_layout() 
```

### Contexto a los gr√°ficos con t√≠tulos
#### Para crear los t√≠tulos realiza lo siguiente:

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b')
ax1.set_title("Relacion X - Y")

ax2.plot(y,x, 'r')
ax2.set_title("Relacion Y - X")
fig.tight_layout() 
```

### Contexto agregando etiquetas
#### Agrega labels en los ejes

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b', )
ax1.set_title("Relacion X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")

ax2.plot(y,x, 'r')
ax2.set_title("Relacion Y - X")
ax2.set_xlabel("Y")
ax2.set_ylabel("X")
fig.tight_layout() 
```

### Cambiar el tama√±o a las figuras
#### Realiza la siguiente acci√≥n:

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize= (1,10))
fig.tight_layout() 
```

### * Aplicar etiquetas (por defecto pone la mejor opci√≥n)

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b', label = "sin(x)")
ax1.set_title("Relacion X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend()
```

### * Modificar la localizaci√≥n de la etiqueta

```python
ax1.legend(loc = "upper center")
```
