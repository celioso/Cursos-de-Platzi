```python
from matplotlib import pyplot as plt
import pandas as pd

# Datos comparativos
data = {
    "Modelo": ["Suzuki Jimny 2017", "Renault Kwid 2025"],
    "Consumo (km/galón)": [43.5, 58],
    "Galones/mes": [22.99, 17.24],
    "Costo mensual (COP)": [367840, 275840]
}

df = pd.DataFrame(data)

# Crear gráfica comparativa
fig, ax = plt.subplots(3, 1, figsize=(10, 12))
fig.suptitle("Comparativa de Consumo y Costo - Jimny vs Kwid", fontsize=16, fontweight='bold')

# Consumo
ax[0].bar(df["Modelo"], df["Consumo (km/galón)"], color=["#1f77b4", "#2ca02c"])
ax[0].set_title("Consumo Promedio (km/galón)")
ax[0].set_ylabel("km/galón")

# Galones mensuales
ax[1].bar(df["Modelo"], df["Galones/mes"], color=["#ff7f0e", "#d62728"])
ax[1].set_title("Consumo Estimado Mensual (galones)")
ax[1].set_ylabel("Galones")

# Costo mensual
ax[2].bar(df["Modelo"], df["Costo mensual (COP)"], color=["#9467bd", "#8c564b"])
ax[2].set_title("Costo Estimado Mensual (COP)")
ax[2].set_ylabel("COP")
ax[2].set_xticklabels(df["Modelo"], rotation=10)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Guardar como imagen
# image_path = "/mnt/data/comparativa_jimny_kwid.png"
image_path = "comparativa_jimny_kwid.png"
plt.savefig(image_path)
image_path
```

```python
# Volver a crear el gráfico incluyendo las etiquetas de valor en cada barra
fig, ax = plt.subplots(3, 1, figsize=(10, 12))
fig.suptitle("Comparativa de Consumo y Costo - Jimny vs Kwid", fontsize=16, fontweight='bold')

# Función auxiliar para agregar etiquetas
def add_labels(ax, values):
    for i, v in enumerate(values):
        ax.text(i, v + (v * 0.02), f"{v:,.2f}", ha='center', fontsize=10, fontweight='bold')

# Consumo
ax[0].bar(df["Modelo"], df["Consumo (km/galón)"], color=["#1f77b4", "#2ca02c"])
ax[0].set_title("Consumo Promedio (km/galón)")
ax[0].set_ylabel("km/galón")
add_labels(ax[0], df["Consumo (km/galón)"])

# Galones mensuales
ax[1].bar(df["Modelo"], df["Galones/mes"], color=["#ff7f0e", "#d62728"])
ax[1].set_title("Consumo Estimado Mensual (galones)")
ax[1].set_ylabel("Galones")
add_labels(ax[1], df["Galones/mes"])

# Costo mensual
ax[2].bar(df["Modelo"], df["Costo mensual (COP)"], color=["#9467bd", "#8c564b"])
ax[2].set_title("Costo Estimado Mensual (COP)")
ax[2].set_ylabel("COP")
ax[2].set_xticklabels(df["Modelo"], rotation=10)
add_labels(ax[2], df["Costo mensual (COP)"])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Guardar como imagen con etiquetas
# image_with_labels_path = "/mnt/data/comparativa_jimny_kwid_etiquetas.png"
image_with_labels_path = "comparativa_jimny_kwid_etiquetas.png"
plt.savefig(image_with_labels_path)
image_with_labels_path
```
