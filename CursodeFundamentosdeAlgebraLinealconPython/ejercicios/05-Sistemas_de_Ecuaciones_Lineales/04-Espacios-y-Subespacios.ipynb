```python
# %matplotlib inline
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
```

```python
v1 = np.array([1,1])
v2 = np.array([-1,-1])
```

```python
for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b,
                    marker='.',
                    color="orange", s =1)

plt.xlim(-25,25)
plt.ylim(-25,25)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

v1 = np.array([1,0,0])
v2 = np.array([2,-3,0])

fig =plt.figure()
ax = fig.add_subplot(111, projection='3d')

for a in range(-10,10):
    for b in range(-10,10):
        ax.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b,
                    v1[2]*a + v2[2]*b,
                    marker='.',
                    color="orange")

ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Vectores en R3
v1 = np.array([1, 0, 0])
v2 = np.array([2, -3, 0])

# Crear figura y ejes 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Dibujar combinaciones lineales de v1 y v2
for a in range(-10, 11):
    for b in range(-10, 11):
        punto = a * v1 + b * v2
        ax.scatter(punto[0], punto[1], punto[2], marker='.', color="orange", alpha=0.6)

# Etiquetas de los ejes
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
ax.set_title('Plano generado por v1 y v2 en R3')

# Mejorar visibilidad del plano
ax.view_init(elev=30, azim=60)  # Cambia perspectiva

plt.tight_layout()
plt.show()
```
