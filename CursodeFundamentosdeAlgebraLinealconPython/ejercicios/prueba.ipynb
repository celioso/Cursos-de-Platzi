```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar dataset
vuelos = sns.load_dataset("flights")

# Reorganizar el DataFrame para que los meses estén como filas y los años como columnas
vuelos = vuelos.pivot(index="month", columns="year", values="passengers")

# Crear mapa de calor
ax = sns.heatmap(vuelos, cmap="YlGnBu", annot=True, fmt="d")

# Mostrar gráfico
plt.title("Mapa de calor de pasajeros por mes y año")
plt.show()
```
