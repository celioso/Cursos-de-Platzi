```python
import numpy as np
import matplotlib.pyplot as plt
```

## sigmoid

```python
def sigmoid(a):
  return 1/(1+np.exp(-a))
```

```python
def step(x):
  return np.piecewise(x, [x < 0.0, x > 0.0],[0, 1])
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def tanh(x):
  return np.sinh(2*x)/(np.cosh(2*x) +1)
```

```python
x = np.linspace(10, -10, 100)
```

```python
plt.plot(x, sigmoid(x))
```

```python
plt.plot(x, step(x))
```

```python
y = relu(x)

# Graficar la función ReLU
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='ReLU(x)', color='blue')
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.axhline(0, color='black', linewidth=0.5, ls='--')  # Línea horizontal en y=0
plt.axvline(0, color='black', linewidth=0.5, ls='--')  # Línea vertical en x=0
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x, tanh(x))
```

```python
def mse(y, y_hat, derivate=False):
  if derivate:
    return (y_hat - y)
  else:
    return np.mean((y_hat - y)**2)
```

```python
real = np.array([0, 0, 1, 1])
prediction = np.array([0.9, 0.5, 0.2, 0.0])
```

```python
mse(real, prediction)
```
